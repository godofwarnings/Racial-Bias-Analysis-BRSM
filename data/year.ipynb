```python
import numpy as np 
import pandas as pd
import glob
import os
from tqdm import tqdm
import regex as re
from collections import defaultdict
import concurrent.futures

import plotly.graph_objects as go
```

```python
def extract_filename(filenames):
    """
    Extracts the filename without extension from a list of filenames with paths.
    
    Args:
        filenames (list): A list of filenames with paths.
        
    Returns:
        list: A list of filenames without extensions.
    """
    filenames_without_ext = []
    for filename in filenames:
        path, filename_with_ext = os.path.split(filename)
        filename, ext = os.path.splitext(filename_with_ext)
        filenames_without_ext.append(filename)
    
    return filenames_without_ext
```

```python
def file_len(file_path):
    with open(file_path, 'r') as file:
        for i, _ in enumerate(file):
            pass
    return i + 1
```

```python
# Initialize a defaultdict to store counts for each year and race
yearly_race_counts = defaultdict(lambda: defaultdict(int))

# Define a function to process each CSV file
def process_csv(file_path):
    df = pd.read_csv(file_path)
    df['date'] = pd.to_datetime(df['date'])
    for index, row in df.iterrows():
        # Extract year from the date column
        year = row['date'].year
        # Increment the count for the corresponding year and race
        yearly_race_counts[year][row['subject_race']] += 1

# Loop through your CSV files and process each one
for file_path in tqdm(all_files, desc='Processing CSV files', unit='file'):
    print("Loading file: ", file_path)
    process_csv(file_path)
    print("Finished processing file: ", file_path)
```

```python
for year, race_counts in yearly_race_counts.items():
    print(f"Year: {year}")
    for race, count in race_counts.items():
        print(f"{race}: {count}")
    print()
```

```python
# Initialize an empty dictionary to store data
data = {}

# Loop through each CSV file
i = 0
for file_name in all_files_pop:
    # Read only 3rd and 4th rows of the CSV file
    df = pd.read_csv(file_name, skiprows=[0, 1], nrows=2, header=None)
    # print(df)
    df.columns = df.iloc[0]
    # Remove the 3rd row
    df = df.iloc[1:]
    # Assign column names as the file names (years)
    year = filenames_pop[i]  # Extract year from file name
    data[year] = df
    i=i+1

# Combine data into a single DataFrame
combined_df = pd.concat(data)

index = combined_df.index
index = [x[0] for x in index]
# convert index to integer
index = [int(x) for x in index]
# print(index)
# set index of the combined_df
combined_df.index = index
# drop columns
combined_df = combined_df.drop(columns=['Footnotes', 'Location'])

# switch row and columns
combined_df = combined_df.T

# print(combined_df.columns)
# reorder the columns in increasing order of column value as integer
combined_df = combined_df[sorted(combined_df.columns)]


# remove name of the index
combined_df.index.name = None
combined_df
```
